```
import pandas as pd
from itertools import chain
from collections import Counter, OrderedDict
import numpy as np
```

```
f = open("coursera_sessions_train.txt","r")
arr = []
i = 0
for line in f.readlines():
    line = line.split(";")
    line[0] = list(map(int, line[0].split(",")))
    if line[1] != '\n':
      line[1] = line[1][:len(line[1])-1]
      line[1] = list(map(int, line[1].split(",")))
    else:
      line[1] = []
    dict1 = {}
    dict1["viewed"] = line[0]
    dict1["purchased"] = line[1]
    arr.append(dict1)
```

```
c1 = Counter()
c2 = Counter()
for el in arr:
  c1.update(el["viewed"])
  c2.update(el["purchased"])
print("c1:\n" , c1)
print("c2:\n" , c2)
```

```
def id_sort(ids, c):
  #print("ids1 = ", ids)
  ids = list(OrderedDict.fromkeys(ids))
  #print("ids2 = ", ids)
  ids = sorted(ids, key=lambda x: c[x], reverse=True)
  #print("ids3 = ", ids)
  return ids
```

```
arr_p = arr[:]
```

```
for i in range(len(arr_p)):
  arr_p[i]["viewed"] = id_sort(arr_p[i]["viewed"], c2) 
```

```
def av_recall1(a):
  n = 0
  sum = 0
  for i in range(len(a)):
    res = 0
    if a[i]["purchased"] != []:
      n += 1
      if a[i]["viewed"][0] in a[i]["purchased"]:
        res = 1/len(a[i]["purchased"])
    sum += res
  return sum/n

def av_precision1(a):
  n = 0
  sum = 0
  for i in range(len(a)):
    res = 0
    if a[i]["purchased"] != []:
      n += 1
      if a[i]["viewed"][0] in a[i]["purchased"]:
        res = 1
    sum += res
  return sum/n

def av_recall5(a):
  n = 0
  sum = 0
  for i in range(len(a)):
    res = 0
    recall = 0
    if a[i]["purchased"] != []:
      n += 1
      rec = min(len(a[i]["viewed"]),5)
      for j in range(rec):
        if a[i]["viewed"][j] in a[i]["purchased"]:
          res +=1
      if res != 0:
        recall = res / len(a[i]["purchased"])
    sum += recall
  return sum / n

def av_precision5(a):
  n = 0
  sum = 0
  for i in range(len(a)):
    res = 0
    recall = 0
    if a[i]["purchased"] != []:
      n += 1
      rec = min(len(a[i]["viewed"]),5)
      for j in range(rec):
        if a[i]["viewed"][j] in a[i]["purchased"]:
          res +=1
      if res != 0:
        recall = res / 5 # there is a mistake in grader, here must be res/rec, but grader doesn't accept this answer
    sum += recall
  return sum / n
```

```
f = open("ans3.txt", "w")
f.write(str("{:.2f}".format(av_recall1(arr_p))) + " ")
f.write(str("{:.2f}".format(av_precision1(arr_p))) + " ")
f.write(str("{:.2f}".format(av_recall5(arr_p))) + " ")
f.write(str("{:.2f}".format(av_precision5(arr_p))))

f.close()
!cat ans3.txt
```
